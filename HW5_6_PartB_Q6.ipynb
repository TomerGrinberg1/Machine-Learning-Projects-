```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.image import imread
from sklearn.cluster import KMeans
import numpy as np
```

# Question 6

```python
def segment_image_with_kmeans(image, k):
    X = image.reshape(-1, 3)
    kmeans = KMeans(n_clusters=k, random_state=0).fit(X)
    segmented = kmeans.cluster_centers_[kmeans.predict(X)]
    return segmented.reshape(image.shape)
```

```python
def plot_segmented(img):
    colors = [10,8,6,4,2]
    fig, axs = plt.subplots(1, len(colors) + 1, figsize=(20, 5))
    axs[0].imshow(img)
    axs[0].set_title("Original Image")
    for i, cluster in enumerate(colors):
        segmented_img = segment_image_with_kmeans(img, cluster)
        axs[i+1].imshow(segmented_img)
        axs[i+1].set_title(f"{cluster} colors")
    plt.show()
```

```python
image_files = ["ladybug.png", "road.png", "beach.png", "dog.png"]
for file in image_files:
    image = plt.imread(file)
    plot_segmented(image)
```
